```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
sentence = "I love this!"
```

```python
inputs = tokenizer(sentence, return_tensors = "pt")  # Can be pyTorch or TensorFlow (tf) object
```

```python
outputs = model(**inputs)
```

```python
probabilities = torch.softmax(outputs.logits, dim = 1)    # dim = 1 for column
probabilities
```

```python
label_id = torch.argmax(probabilities, dim = 1).item()
label = model.config.id2label[label_id]
label_id
```

```python
print("Sentence: ", {sentence})
print("Sentiment: ", {label})
print(f"Probabilities: , {probabilities.tolist()}")
```
