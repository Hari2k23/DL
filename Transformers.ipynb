```python
from transformers import pipeline
```

```python
# Sentiment Analysis
classifier = pipeline("sentiment-analysis")
classifier("This is the best day of my life.")
classifier(['I love this so much', 'I hate this so much', 'A normal day'])
```

```python
# Zero Shot Classification
classifier = pipeline("zero-shot-classification")
classifier(
    "Vijay started his new party named TVK on Feb 2024",
    candidate_labels = ["education", "politics", "business"]
)
```

```python
# Text Generation
generator = pipeline("text-generation", model = "distilgpt2")
generator("AI in the future will be", max_length = 10)
```

```python
# Fill Masking
unmasker = pipeline("fill-mask")
unmasker("This course is about <mask> models.", top_k = 5)
```

```python
# NER
ner = pipeline("ner", grouped_entities = True)
ner("His name is Virat Kohli and he is a cricketer of India, born on 5th November 1988")
```
