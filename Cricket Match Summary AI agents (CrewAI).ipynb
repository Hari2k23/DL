```python
from crewai import Agent, Task, Crew
from langchain_community.chat_models import ChatLiteLLM
import os

os.environ["GOOGLE_API_KEY"] = "AIzaSyAWQy36sURjDNvonE-tAfH95F4zlA71xHM"

llm = ChatLiteLLM(
    model="gemini/gemini-1.5-flash",
    temperature=0,
    api_key=os.environ["GOOGLE_API_KEY"]
)

news_extractor = Agent(
    role="News Extractor",
    goal="Gather match data like scorecards and highlights",
    backstory="A sports journalist with expertise in cricket stats and updates.",
    verbose=True,
    llm=llm
)

stat_analyst = Agent(
    role="Statistical Analyst",
    goal="Analyze player performances and match statistics",
    backstory="An experienced cricket analyst skilled in stats like strike rate, wickets, and run rate.",
    verbose=True,
    llm=llm
)

fan_voice = Agent(
    role="Fan Voice",
    goal="Write emotional fan reactions about the match",
    backstory="A die-hard cricket fan who shares passionate commentary reflecting the fans' feelings.",
    verbose=True,
    llm=llm
)

match_reporter = Agent(
    role="Match Reporter",
    goal="Create a final smart report combining stats, news, and fan emotions",
    backstory="A senior editor who creates polished match summaries by combining data and emotions.",
    verbose=True,
    llm=llm
)

news_task = Task(
    description="""
Gather key match data from the India vs Pakistan T20 World Cup 2022 match at MCG.
Include final scores, top performers, fall of wickets, and notable highlights.
""",
    agent=news_extractor,
    expected_output="Scorecard, match highlights, and list of top-performing players."
)

stat_task = Task(
    description="""
Analyze the match performances.
Mention top scorers, strike rates, top wicket takers, and turning points of the match.
""",
    agent=stat_analyst,
    context=[news_task],
    expected_output="Statistical breakdown of individual and team performances."
)

fan_text = """
They are jumping on to Kohli. He has been engulfed by his mates. Rohit Sharma has lifted him on his arms and twirling him like a child. Is this the best T20 knock when there is so much of pressure? This is stuff of legends. It happens in fables and Bollywood potboilers when the hero comes out of nowhere to save the day. What a man! What a legend! Kohli seemed to be a little teary-eyed. India were dead and buried at 31 for 4. But Kohli and Pandya saw a few overs off. Then went for the attack after 10 overs and have managed to script something historic. But credit to Pakistan. They were ruffled early on but came back with the bat first in the second half and then carried the momentum when they had the ball in their hands. The seamers turned on the heat and Rohit and Rahul had no answers. SKY fizzled out too soon. Axar couldn't do what Nawaz did earlier in the Asia Cup.

Indian fans have heard folklores of Miandad finishing it off with a sixer in the late 80s. Then Afridi did the same in 2014. But in a close chase, Kohli and Pandya have managed to come up with something extraordinary. The partnership of Kohli-Pandya ran hard and Kohli managed to clear the long fences four times. Three of those came when it seemed it might well be over for India. The broadcasters are still struggling to talk amongst themselves. Not a single soul has left the stadium. Can there be a better way to market this sport? A massive section of cricket fans had given up on Kohli. But he came back with renewed vigour during Asia Cup and today's knock is a gentle reminder than form is temporary and class is permanent!

And when he hit that straight six off Haris Rauf — standing tall, perfectly still — it was like the clock froze. Time stood still for a moment. Every fan from Wankhede to Melbourne screamed in unison. That shot will live forever — like Sachin’s upper cut, Dhoni’s final six, and Yuvi’s six sixes.

This wasn’t just a win. This was emotion. Redemption. And a reminder that even when the chips are down, Virat Kohli is never out. The King is back.
"""

fan_task = Task(
    description=f"""
Write a passionate fan commentary for India vs Pakistan T20 World Cup 2022 match at MCG.
Use the following emotions and text as inspiration:\n\n{fan_text}
""",
    agent=fan_voice,
    expected_output="Emotional and energetic fan commentary."
)

final_report_task = Task(
    description="""
Create a complete match report combining:
- News and stats
- Analytical points
- Emotional commentary

Ensure it's detailed, structured, and engaging for readers.
""",
    agent=match_reporter,
    context=[news_task, stat_task, fan_task],
    expected_output="A final smart summary report of the match."
)

crew = Crew(
    agents=[news_extractor, stat_analyst, fan_voice, match_reporter],
    tasks=[news_task, stat_task, fan_task, final_report_task],
    verbose=True
)

result = crew.kickoff()
print("\nFinal Cricket Match Summary:\n")
print(result)
```
