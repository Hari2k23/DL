```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
# from tensorflow.keras.models import Model
# from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical # One-hot encoding
```

```python
(xtrain, ytrain), (xtest, ytest) = datasets.mnist.load_data()

xtrain = xtrain.reshape((60000, 28, 28)).astype('float32')/255   # Reshaping and normalizing
xtest = xtest.reshape((10000, 28, 28)).astype('float32')/255

ytrain = to_categorical(ytrain)
ytest = to_categorical(ytest)
```

```python
print(xtrain.shape, ytrain.shape, xtest.shape, ytest.shape)
```

```python
input_shape = (28, 28)
inputs = layers.Input(shape = input_shape)
rnn_layer = layers.LSTM(128)(inputs)
outputs = layers.Dense(10, activation = 'softmax')(rnn_layer)

model = models.Model(inputs = inputs, outputs = outputs)
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(xtrain, ytrain, validation_data = (xtest, ytest), epochs = 10, batch_size = 64)
```

```python
test_loss, test_accuracy = model.evaluate(xtest, ytest)
print("Test Accuracy:", test_accuracy)
```
