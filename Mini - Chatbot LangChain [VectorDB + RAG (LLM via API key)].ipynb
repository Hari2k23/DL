```python
from langchain.document_loaders import TextLoader            # To load doc file
from langchain.text_splitter import CharacterTextSplitter    # To input as chunks
from langchain.embeddings import HuggingFaceEmbeddings       # Word embeddings
from langchain.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA
import os
```

```python
loader = TextLoader("sample.txt")
documents = loader.load()
```

```python
splitter = CharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
docs = splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings()
vectorDB = FAISS.from_documents(docs, embeddings)
```

```python
# Given as environment variable, hence store panna theva illa
os.environ["GOOGLE_API_KEY"] = "AIzaSyAWQy36sURjDNvonE-tAfH95F4zlA71xHM"
llm = ChatGoogleGenerativeAI(
    model =  "gemini-1.5-flash", temperature = 0      # temperature - for creativity
)

qa_chain = RetrievalQA.from_chain_type(
    llm = llm, retriever = vectorDB.as_retriever()
)

question = "Who is MSD?"
answer = qa_chain.run(question)

print("Answer:\n", answer)
```
